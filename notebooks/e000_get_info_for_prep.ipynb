## 結論
 - 読み込みが早くて圧縮率も良い gzip で良さそう
 - parquet 使ったら書き込み読み込み共に早いので良さそう (float16) に対応していないのが面倒だが
 - 今回のコンペでは read, write 共にそんなに問題なさそうなので取りあえず pkl.gz でいく
